```python
import numpy as np
import pandas as pd
import scipy.io as scio

%load_ext autoreload
%autoreload 2
import main
```

```python
mat = scio.loadmat('ex6data1.mat')

x = pd.DataFrame(mat['X'], columns=['x1', 'x2'])
```

```python
K = 3
iterations = 30

progress = list(main.kmeans_multi(x, K, iterations))
clusters, centroids = progress[-1]
cost = main.cost(x, clusters, centroids)
print('Cost:', cost)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

size = (3, len(progress) * 3)
_, subplots = plt.subplots(len(progress), 1, figsize=size)
for (clusters, centroids), ax in zip(progress, subplots):
    for k in range(K):
        ax.scatter(x.x1, x.x2, c=clusters)
    ax.scatter(centroids.iloc[:, 0], centroids.iloc[:, 1], s=100, c='red', marker='+')

plt.show()
```

```python
import numpy as np
import pandas as pd
import scipy.io as scio
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

%load_ext autoreload
%autoreload 2
import main
```

```python
mat = scio.loadmat('ex7data1.mat')
x = pd.DataFrame(mat['X'])

scaler = StandardScaler()
scaled = scaler.fit_transform(x)
```

```python
projections, uk, _ = main.pca(scaled, 2)

plt.scatter(scaled[:, 0], scaled[:, 1])
plt.arrow(0, 0, uk[0, 0], uk[0, 1], color='red', head_width=0.2)
plt.arrow(0, 0, uk[1, 0], uk[1, 1], color='red', head_width=0.1)

plt.show()
```

```python
projections, uk, _ = main.pca(scaled, 1)
inv = main.pca_inverse(projections, uk)

plt.scatter(inv[:, 0], inv[:, 1])
plt.arrow(0, 0, uk[0, 0], uk[0, 1], color='red', head_width=0.1)
plt.show()
```
